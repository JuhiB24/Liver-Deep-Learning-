```python
import tensorflow as tf

# Load the model using TensorFlow's built-in method
model = tf.keras.models.load_model('D:/PATENT/Model_testing/steatosis_modified15.h5')
```

```python
# plotting neural network of the model
import PlotNeuralNet
```

```python
import tensorflow as tf
from PlotNeuralNet import PlotNeuralNet

# Extract the model's architecture
model_layers = model.layers

# Define the neural network architecture using PlotNeuralNet syntax
network_architecture = []

# Add input layer
network_architecture.append({"layer": "Input", "shape": model.input_shape[1:], "name": "input"})

# Add remaining layers
for layer in model_layers:
    if isinstance(layer, tf.keras.layers.Conv2D):
        network_architecture.append({
            "layer": "Conv2D",
            "filters": layer.filters,
            "kernel_size": layer.kernel_size,
            "strides": layer.strides,
            "activation": layer.activation.__name__,
            "name": layer.name
        })
    elif isinstance(layer, tf.keras.layers.MaxPooling2D):
        network_architecture.append({
            "layer": "MaxPooling2D",
            "pool_size": layer.pool_size,
            "strides": layer.strides,
            "name": layer.name
        })
    elif isinstance(layer, tf.keras.layers.Dense):
        network_architecture.append({
            "layer": "FullyConnected",
            "units": layer.units,
            "activation": layer.activation.__name__,
            "name": layer.name
        })

# Generate the neural network diagram
PlotNeuralNet(model, "D:/PATENT/JUHI_22-10-24/Fibrosis_architechture.png")
```
